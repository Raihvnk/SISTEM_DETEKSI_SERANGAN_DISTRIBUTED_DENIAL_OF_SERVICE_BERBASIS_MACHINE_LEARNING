```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graph kosong
G = nx.Graph()

# Menambahkan node dengan label
G.add_node('h1 (Victim)', color='red')
G.add_node('h2 (Attacker)', color='blue')
G.add_node('h3 (Attacker)', color='blue')
G.add_node('h4 (Attacker)', color='blue')
G.add_node('h5 (Normal)', color='green')
G.add_node('h6 (Normal)', color='green')
G.add_node('s1 (Switch)', color='grey')

# Menambahkan edge untuk menghubungkan node
G.add_edges_from([
    ('h1 (Victim)', 's1 (Switch)'),
    ('h2 (Attacker)', 's1 (Switch)'),
    ('h3 (Attacker)', 's1 (Switch)'),
    ('h4 (Attacker)', 's1 (Switch)'),
    ('h5 (Normal)', 's1 (Switch)'),
    ('h6 (Normal)', 's1 (Switch)')
])

# Mengambil warna untuk node berdasarkan peran
color_map = [G.nodes[node]['color'] for node in G]

# Visualisasi topologi
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color=color_map, node_size=2000, font_size=10, font_color='white', font_weight='bold')
plt.title('SDN DDoS Attack Topology')
plt.show()
```
